```python
#按照前兩個index去做padding，由於最長句子是777直接填777
def sequence_padding(df, max_length, dim, zero=0):
    multi_index = pd.unique([ (mi[0], mi[1]) for mi in df.index ])
    padding = list()
    for mi in multi_index:
        m = df.ix[mi].as_matrix()
        b = np.array([zero]*dim)
        
        while m.shape[0] < max_length:
            b = b.reshape(1, dim)
            m = np.concatenate((m, b), axis=0)
        else:
            padding.append(m)
    
    return np.asarray(padding)
```

```python
mfcc_train_padding = sequence_padding(mfcc_train_multi_index, 777, 39)
mfcc_test_padding = sequence_padding(mfcc_test_multi_index, 777, 39)
fbank_train_padding = sequence_padding(fbank_train_multi_index, 777, 69)
fbank_test_padding = sequence_padding(fbank_test_multi_index, 777, 69)
mfcc_fbank_train_padding = sequence_padding(mfcc_fbank_train_multi_index, 777, 108)
mfcc_fbank_test_padding = sequence_padding(mfcc_fbank_test_multi_index, 777, 108)
```

```python
#直接機將所有訊號當成連續的，平均frame數量是304，將每個都切成304(只有最後一個不足304捕到304筆)，但效果很差(不確定有沒有處理正確)
def slice_padding(df, dim, zero=0, slice_num = 304):
    df = df.as_matrix()
    padding = list()
    b = np.array([zero]*dim)
    b = b.reshape(1, dim)
    print(b.shape)
    
    for sta in list(range(0, df.shape[0], slice_num)):
        if sta+slice_num > df.shape[0]:
            m = df[sta:]
            while m.shape[0] < slice_num:
                m = np.concatenate((m, b), axis=0)
        else:
            m = df[sta:sta+slice_num]
        
        padding.append(m)
    return np.asarray(padding)
```

```python
# 處理mapping之間的轉換
# phone_to_number是48個pnohne對到48個數字
# number_to_phone上面的inversion mapping
# number_to_char 48個數字map到char
# number_40_to_char_39 則是直接將48個輸出對應到39個字母

with open(data_dir+'48phone_char.map') as phone_map:
    phone_to_number = dict()
    number_to_phone = dict()
    number_to_char = dict()
    for pm in phone_map:
        (a, b, c) = pm.replace('\n','').split('\t')
        phone_to_number[a] = int(b)
        number_to_phone[int(b)] = a
        number_to_char[int(b)] = c
        
with open(data_dir+'phones/48_39.map') as phone_map:
    phone_48_to_39 = dict()
    for pm in phone_map:
        (a, b) = pm.replace('\n','').split('\t')
        phone_48_to_39[a] = b  
        
number_48_to_39 = dict()
number_48_to_char_39 = dict()
for n48, v48 in number_to_phone.items():
    v39 = phone_48_to_39[v48]
    n39 = phone_to_number[v39]
    c = number_to_char[n39]
    #print(n48, v48, v39, n39, c)
    number_48_to_39[n48] = n39
    number_48_to_char_39[n48] = c
```

```python

# 處理最後output用的，將重複tag拿掉
def compact(list_in):
    #list_out = list()
    e0 = list_in[0]
    list_out = [e0]
    for e1 in list_in[1:]:
        if e1 == e0:
            continue
        else:
            e0 = e1
            list_out.append(e0)
        
    return list_out

# 處理最後output用的，將頭尾sil去掉，重複的tag拿掉，並轉為字母
def out_csv(pred, prefix):
    sample = pd.DataFrame.from_csv('sample.csv')
    test_index = list(sample.index)
    with open(prefix+'.csv', 'w') as csv:
        
        csv.write('id,phone_sequence\n')
        for p, t in zip(pred, test_index):
            pc = compact(p)
            if pc[0] == 37:
                del pc[0]
            if pc [-1] == 37:
                del pc[-1]
            #print(''.join([number_48_to_char_39[x] for x in pc]))
            csv.write(t+','+''.join([number_48_to_char_39[x] for x in pc])+'\n')
```
